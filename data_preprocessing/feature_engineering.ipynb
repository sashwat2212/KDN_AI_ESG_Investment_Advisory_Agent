```python
import pandas as pd

df = pd.read_csv("/Users/kdn_aisashwat/Desktop/esg-investment-advisor /final_esg_dataset.csv")
```

```python
df['3_month_MA'] = df['Total_Score'].rolling(window=3).mean()
df['6_month_MA'] = df['Total_Score'].rolling(window=6).mean()
```

```python
df['ESG_Change_3m'] = df['Total_Score'].pct_change(periods=3)
df['ESG_Change_6m'] = df['Total_Score'].pct_change(periods=6)
```

```python
df['Env_Score_Lag_1'] = df['Env_Score'].shift(1)
df['Soc_Score_Lag_1'] = df['Soc_Score'].shift(1)
```

```python
df['Weighted_ESG_Score'] = (df['Env_Score'] * 0.4) + (df['Soc_Score'] * 0.3) + (df['Gov_Score'] * 0.3)
```

```python
df['Forest_Cover_Change'] = df['avg_deforestation'].pct_change(periods=6)
```

```python
df['Greenwashing_Risk'] = ((df['Total_Score'] > 75) & (df['twitter_esg_sentiment'] < 0.5)).astype(int)
```

```python
df
```

```python
df.drop(columns=[
    'Date', 
    'Forest_Cover_Change'
], inplace=True)
```

```python
df.to_csv('final_processed_esg_data.csv', index=False)
```

```python
from sklearn.preprocessing import MinMaxScaler

#normalize
columns_to_normalize = [
    "avg_deforestation",
    "avg_air_pollution",
    "avg_water_quality_B3",
    "avg_water_quality_B4",
    "avg_water_quality_B5"
]

#median
df[columns_to_normalize] = df[columns_to_normalize].fillna(df[columns_to_normalize].median())

#min-max scaling
scaler = MinMaxScaler()
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])


df.to_csv("normalized_final_esg_data.csv", index=False)

print(" Data normalization complete! Saved as 'normalized_final_esg_data.csv'.")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path)

    #filling withforward fill
    env_cols = ['avg_deforestation', 'avg_air_pollution', 
                'avg_water_quality_B3', 'avg_water_quality_B4', 'avg_water_quality_B5']
    df[env_cols] = df[env_cols].fillna(method='ffill')

    #normalize
    scaler = MinMaxScaler()
    df[env_cols] = scaler.fit_transform(df[env_cols])

    return df
```
